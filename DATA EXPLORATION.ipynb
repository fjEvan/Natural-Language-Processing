```python
def percentstandardize_barplot(x,y,hue, data, ax=None, order= None):
    """
    Standardize by percentage the data using pandas functions, then plot using Seaborn.
    Function arguments are and extention of Seaborns'.
    """
    sns.barplot(x= x, y=y, hue=hue, ax=ax, order=order,
    data=(data[[x, hue]]
     .reset_index(drop=True)
     .groupby([x])[hue]
     .value_counts(normalize=True)
     .rename('Percentage').mul(100)
     .reset_index()
     .sort_values(hue)))
    plt.xlabel("Rating",fontsize=16)
    plt.title("Occurrence of Recommended/Not by Rating",fontsize=16)
    plt.ylabel("% Percentage by Rating",fontsize=16)
    plt.xticks(fontsize=18)
    plt.yticks(fontsize=18)
    plt.show()
    plt.savefig('pf.jpg')
import seaborn as sns
huevar = "Rating"
plt.figure(figsize=(10,8))
percentstandardize_barplot(x="Rating",y="Percentage", hue="Recommended IND",data=df)


```
